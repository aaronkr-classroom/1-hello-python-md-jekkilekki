```python
# Numpy = array library
import numpy as np

# Matplotlib = plotting library
import matplotlib.pyplot as plt

# Core library
import math
```

```python
# 간단한 배열 만들기
t = np.linspace(start = 0, stop = 2*math.pi, num = 100, dtype = np.float32)
y = np.sin(t)
```

```python
# 간단한 표
plt.figure()
plt.plot(t,y)
plt.xlabel('Time')
plt.ylabel('Signal')
plt.show()
```

```python
# 산포도

# 3차원에서 100개 난수 만들기
x, y, scale = np.random.randn(3, 100)
fig, ax = plt.subplots()

# 산포도 표시하기
ax.scatter(x = x, y = y, c = scale, s = np.abs(scale) * 500)
ax.set(title = "Random data in a scatterplot")
plt.show()
```
